```python
import pandas as pd
```

```python
df = pd.read_csv('../data/persuade-predictions-long.csv')
models = [x.split('_')[0] for x in list(df['model'].drop_duplicates())]
models
```

```python
for model in models:
    df['cons_'+model] = df['model'].apply(lambda x: 1 if x == model+'_pred' else 0)
    for i in list(df['race_ethnicity'].drop_duplicates()):
        df['cons_'+i] = df.apply(lambda row: row['race_ethnicity'] if (row['model'] == model+'_pred') & (row['race_ethnicity'] == i) else 0, axis=1)
df
```
